3)	How many players have hit 40 or more HRs in one single season? (Number only)

```python
new_df=df.groupby(['playerID','yearID'],as_index=False).sum()
new_df[new_df['HR']>=40]['playerID'].nunique()
```

4)	How many players have hit 600 or more HRs for their career? (Dataframe with name and player first and last name only and amount of HRs)

```python
new_df=df.groupby(['playerID','nameFirst','nameLast'],as_index=False).sum()
new_df[new_df['HR']>=600].loc[:,['nameFirst','nameLast','HR']].sort_values('HR',ascending=False)
```

5)	How many unique players have hit 40 2Bs or more, 10 3Bs or more, 200 Hits or more, and 30 HRs or more in one season? (Number Only)

```python
new_df=df.groupby(['playerID','yearID'],as_index=False).sum()
new_df[(new_df['2B']>=40)&(new_df['3B']>=10)&(new_df['H']>=200)&(new_df["HR"]>=30)]['playerID'].nunique()
```

6)	How many players seasons have had 100 or more SBs? (Dataframe, first name, last name, yearID, number of SBs, ordered from descending SBs )

```python
new_df=df.groupby(['playerID','nameFirst','nameLast','yearID'],as_index=False).sum()
new_df[new_df['SB']>=100].loc[:,['nameFirst','nameLast','yearID','SB']].sort_values(by=['SB'],ascending=False)
```

7)	How many players in the 1960s (yearID) have hit 200 or more HRs? (Dataframe, first name, last name, number of HRs, ordered descending by HR amt)

8)	Who has hit the most HRs in history? (Dataframe, first name, last name, HRs)

9)	Who had the most hits (H) in the 1970s? (Dataframe)

```python
new_df=df[(df['yearID']>=1970)&(df['yearID']<1980)].groupby(['playerID','nameFirst','nameLast'],as_index=False).sum()
new_df[new_df['H']==new_df['H'].max()].loc[:,['nameFirst','nameLast','H']]
```

10)	Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  (Dataframe, first name, last name, OBP, ordered descending by OBP)

```python
new_df=df[df['yearID']==1977].groupby(['playerID','nameFirst','nameLast'],as_index=False).sum()
new_df['OBP']=(new_df['H']+new_df['BB']+new_df['HBP'])/(new_df['AB']+new_df['BB']+new_df['HBP']+new_df['SF'])
new_df['PA']=new_df['AB']+new_df['BB']+new_df['HBP']+new_df['SH']+new_df['SF']
new_df[new_df['PA']>=500].loc[:,['nameFirst','nameLast','OBP']].sort_values(by=['OBP'],ascending=False).head(5)
```

11)	Top 8 highest averages in 2013 with at least 300 PAs? (Dataframe, first name, last name, average, descending by average)

```python
new_df=df[df['yearID']==2013].groupby(['playerID','nameFirst','nameLast'],as_index=False).sum()
new_df['PA']=new_df['AB']+new_df['BB']+new_df['HBP']+new_df['SH']+new_df['SF']
new_df['Average']=new_df['H']/new_df['AB']
new_df[new_df['PA']>=300].loc[:,['nameFirst','nameLast','Average']].sort_values(by=['Average'],ascending=False).head(8)
```

23)	Who had more than the league average HRs (for players with 500 or more ABs) in 2012 (filter out all players with less 500 ABs)? (Dataframe first name, last name, HR descending)

```python
new_df=df[(df['yearID']==2012)].groupby(['playerID','nameFirst','nameLast','lgID'],as_index=False).sum()
new_df['Avg']=new_df[new_df['AB']>=500].groupby('lgID')['HR'].transform('mean')
new_df[new_df['HR']>new_df['Avg']][['nameFirst','nameLast','HR']].sort_values(by=['HR'],ascending=False)
```

24)	 Who is the youngest player to hit 50 or more HRs in a single season? (Dataframe, first name, last name, HRs, season)

```python
new_df=df.groupby(['yearID','birthYear','playerID','nameFirst','nameLast'],as_index=False).sum()
new_df['Age']=new_df['yearID']-new_df['birthYear']
df1=new_df[new_df['HR']>=50]
df1[df1['Age']==df1['Age'].min()][['nameFirst','nameLast','HR','yearID','Age']]
```
